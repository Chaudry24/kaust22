```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("dark_background")
```

```python
# LOAD TRAINING DATASETS
data_train_1 = pd.read_csv("./data-competition-1a/1a_1_train.csv")
data_train_2 = pd.read_csv("./data-competition-1a/1a_2_train.csv")

# LOAD TESTING DATASETS
data_test_1 = pd.read_csv("./data-competition-1a/1a_1_test.csv")
data_test_2 = pd.read_csv("./data-competition-1a/1a_2_test.csv")
```

```python
# CONVERT DATASET 1 TO NUMPY
train_x1 = pd.DataFrame.to_numpy(data_train_1)[:, [1, 2]]
train_y1 = pd.DataFrame.to_numpy(data_train_1)[:, -1]

# CONVERT DATASET 2 TO NUMPY
train_x2 = pd.DataFrame.to_numpy(data_train_2)[:, [1, 2]]
train_y2 = pd.DataFrame.to_numpy(data_train_2)[:, -1]

# CONVERT TESTSETS TO NUMPY
test_obs1 = pd.DataFrame.to_numpy(data_test_1)[:, [1, 2]]
test_obs2 = pd.DataFrame.to_numpy(data_test_2)[:, [1, 2]]
```

```python
# VISUALIZE DATA
plt.figure()
plt.title("Data 1")
plt.scatter(x=train_x1[:, 0], y=train_x1[:, 1], c=train_y1, cmap="jet",
           s=1)
plt.colorbar()

plt.figure()
plt.title("Data 2")
plt.scatter(x=train_x2[:, 0], y=train_x2[:, 1], c=train_y2, cmap="jet",
           s=1)
plt.colorbar()
```
